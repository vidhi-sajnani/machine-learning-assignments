```python
import pandas as pd
```

```python
#printing the data
header_names = ['YEAR','Rainfall in June']   
data = pd.read_csv('data.csv', usecols=header_names) 
data
```

```python
#imputing missing values with median
data['Rainfall in June']= data['Rainfall in June'].fillna(data["Rainfall in June"].median())
#data['Rainfall in June']= data['Rainfall in June'].fillna(data["Rainfall in June"].mean())
#data['Rainfall in June'] = data['Rainfall in June'].fillna(data['Rainfall in June'].mode()[0])
data #displaying data
```

```python
#removing duplicates
data=data.drop_duplicates()
data #displaying data, 2 rows removed
```

```python
import matplotlib.pyplot as plt
```

```python
#histogram with the mean and median displayed
plt.hist(data["Rainfall in June"])
plt.xlabel("Rainfall in June")
plt.ylabel("Frequency")
plt.axvline(x = data["Rainfall in June"].mean(), color = 'r')
plt.axvline(x = data["Rainfall in June"].median(), color = 'b')
#we can see that the symmetry is skewed to the left
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
#standard scaler
#z = (x - mean) / standard deviation
model1 = StandardScaler()
ss_data= model1.fit_transform(data)
print(ss_data)
```

```python
#min max scaler
#x_scaled = (x - min) / (max - min)
model2 = MinMaxScaler()
mm_data= model2.fit_transform(data)
print(mm_data)
```

```python
#Differences in range
range=mm_data-ss_data
print(range)
```

```python
#Linear Regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X= data.drop(columns=['Rainfall in June'])
y=data['Rainfall in June']
```

```python
X #independent variable
```

```python
y #dependent variable
```

```python
#training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
model = LinearRegression().fit(X_train,y_train)
```

```python
#scatter plot
plt.scatter(X_train, y_train, color ='b')
plt.plot(X_test, model.predict(X_test), color ='r')
 
plt.show()
```

```python
#printing the metrics
import sklearn.metrics as metrics
import numpy as np
mse1= metrics.mean_squared_error(y_test,model.predict(X_test))

print("Mean Squared Error (MSE) of the test data: ", metrics.mean_squared_error(y_test,model.predict(X_test)))
print("Mean Absolute Error (MAE) of the test data: ", metrics.mean_absolute_error(y_test,model.predict(X_test)))
print("Root Mean Squared Error (RMSE) of the test data: ", np.sqrt(mse1))
```

```python
#using 70:30 split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
model2 = LinearRegression().fit(X_train,y_train)
```

```python
#scatter plot
plt.scatter(X_train, y_train, color ='b')
plt.plot(X_test, model2.predict(X_test), color ='r')
 
plt.show()
```

```python
#printing the metrics
import sklearn.metrics as metrics
import numpy as np
mse2= metrics.mean_squared_error(y_test,model2.predict(X_test))

print("Mean Squared Error (MSE) of the test data: ", metrics.mean_squared_error(y_test,model2.predict(X_test)))
print("Mean Absolute Error (MAE) of the test data: ", metrics.mean_absolute_error(y_test,model2.predict(X_test)))
print("Root Mean Squared Error (RMSE) of the test data: ", np.sqrt(mse2))
```
